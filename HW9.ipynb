### 136. Single Number

Given a non-empty array of integers nums, every element appears twice except for one. 

Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

Example 1:

```
Input: nums = [2,2,1]
Output: 1
```

Example 2:

```
Input: nums = [4,1,2,1,2]
Output: 4
```

Example 3:

```
Input: nums = [1]
Output: 1
``` 


<div class="alert alert-block alert-info">
<b>Constraints:</b>
    
- $1 <= nums.length <= 3 * 10^{4}$
    
- $- 3 * 10^{4} <= nums[i] <= 3 * 10^{4}$
    
- Each element in the array appears twice except for one element which appears only once.
    
</div>

````python
def single_number(nums: list[int]) -> int:
    for i, val in enumerate(nums):
        if val not in nums[:i] + nums[i+1:]:
            return val


def test_single_number():
    assert single_number([2, 2, 1]) == 1
    assert single_number([4, 1, 2, 1, 2]) == 4
    assert single_number([1]) == 1


test_single_number()
````

### 242. Valid Anagram

Given two strings s and t, return true if t is an anagram of s, and false otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase,

typically using all the original letters exactly once.

Example 1:

```
Input: s = "anagram", t = "nagaram"
Output: true
```

Example 2:

```
Input: s = "rat", t = "car"
Output: false
```

<div class="alert alert-block alert-info">
<b>Constraints:</b>
    
- $1 <= s.length, t.length <= 5 * 10^{4}$
    
- s and t consist of lowercase English letters.
    
</div>

````python
def is_anagram(s: str, t: str) -> bool:
    return sorted(s) == sorted(t)


def test_is_anagram():
    # Має бути 1 -> True, 0 -> False, так записано, щоб
    # у flake8-nb не виникало зауваження E712 (==True)
    assert is_anagram(s="anagram", t="nagaram") == 1
    assert is_anagram(s="rat", t="car") == 0


test_is_anagram()
````

### 414. Third Maximum Number

Given an integer array $nums$, return the ***third distinct maximum number*** in this array. 

If the third maximum does not exist, return the ***maximum*** number.

Example 1:

```
Input: nums = [3,2,1]
Output: 1
Explanation:
The first distinct maximum is 3.
The second distinct maximum is 2.
The third distinct maximum is 1.
```

Example 2:

```
Input: nums = [1,2]
Output: 2
Explanation:
The first distinct maximum is 2.
The second distinct maximum is 1.
The third distinct maximum does not exist, so the maximum (2) is returned instead.
```

Example 3:

```
Input: nums = [2,2,3,1]
Output: 1
Explanation:
The first distinct maximum is 3.
The second distinct maximum is 2 (both 2's are counted together since they have the same value).
The third distinct maximum is 1.
``` 


<div class="alert alert-block alert-info">
<b>Constraints:</b>
    
- $1 <= nums.length <= 10^{4}$
    
- $- 2 * 10^{31} <= nums[i] <= 3 * 10^{31}$
    
- Each element in the array appears twice except for one element which appears only once.
    
</div>

**Follow up:** Can you find an $O(n)$ solution?

````python
def third_max(nums: list[int]) -> int:
    '''
    nums = set(nums)
    if len(nums) >= 3:
        return sorted(nums, reverse=True)[2]
    return max(nums)
    '''
    if len(nums) < 3:
        return max(nums)
    i = 3
    while i:
        max_i = max(nums)
        while max_i in nums:
            nums.remove(max_i)
        i -= 1
    return max_i


def test_third_max():
    assert third_max([3, 2, 1]) == 1
    assert third_max([1, 2]) == 2
    assert third_max([2, 2, 3, 1]) == 1


test_third_max()
````

$O(n) = n$

Оскільки, щоб знайти максимальне значення, потрібно перебрати всі n елеметів,

навіть якщо робити це тричі, складність алгоритму не зміниться, бо 

$O(f(n) + a) = O(f(n) * a) = O(f(n))$, де $a = const$
